```python
import matplotlib.pyplot as plt

# 'N (명)' 행 제거
kdrama_copy2 = kdrama_copy2[kdrama_copy2['매체별'] != 'N (명)']

# 연령별 TV와 스마트폰 사용 그래프
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# TV 사용 그래프
tv_data = kdrama_copy2[kdrama_copy2['매체별'].str.contains('TV')].set_index('매체별').T
tv_data.plot(ax=axes[0], marker='o')
axes[0].set_title('연령별 TV 사용')
axes[0].set_ylabel('값')

# 스마트폰 사용 그래프
smartphone_data = kdrama_copy2[kdrama_copy2['매체별'].str.contains('스마트폰')].set_index('매체별').T
smartphone_data.plot(ax=axes[1], marker='o')
axes[1].set_title('연령별 스마트폰 사용')
axes[1].set_xlabel('연도')
axes[1].set_ylabel('값')

plt.tight_layout()
plt.show()
```

```python
# 'N (명)' 행 제거
kdrama_copy2 = kdrama_copy2[kdrama_copy2['매체별'] != 'N (명)']

# '-' 값을 NaN으로 대체
kdrama_copy2.replace('-', np.nan, inplace=True)

# TV 사용 그래프
tv_data = kdrama_copy2[kdrama_copy2['매체별'].str.contains('TV')].set_index('매체별').T
plt.figure(figsize=(10, 5))
plt.plot(tv_data.iloc[::-1], marker='o')
plt.title('연령별 TV 사용')
plt.xlabel('연도')
plt.ylabel('값')
plt.legend(tv_data.columns, fontsize='small', loc='upper right')
plt.ylim(0, 100)
plt.show()
```

```python
# 'N (명)' 행 제거
kdrama_copy2 = kdrama_copy2[kdrama_copy2['매체별'] != 'N (명)']

# '-' 값을 NaN으로 대체
kdrama_copy2.replace('-', np.nan, inplace=True)

# TV 사용 그래프
tv_data = kdrama_copy2[kdrama_copy2['매체별'].str.contains('TV')].set_index('매체별').T

# 70세 이상 데이터 제외
tv_data = tv_data.drop(columns='70세이상_TV')

plt.figure(figsize=(10, 5))

# 각 선에 대한 색상 설정
colors = ['#c2e8ff', '#38b3ff', '#89cffa', '#009eff', '#cf3232', '#ff0000']

# 그래프 그리기
for i, col in enumerate(tv_data.columns):
    if i < len(colors):  # colors 리스트의 범위를 벗어나지 않도록 조건 추가
        plt.plot(tv_data.index, tv_data[col], marker='o', label=col, color=colors[i])

plt.title('연령별 TV 사용')
plt.xlabel('연도')
plt.ylabel('값')
plt.legend(fontsize='small', loc='upper right')

# y축 반전
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

plt.ylim(0, 100)

plt.grid()

plt.savefig('tv_usage_graph.png', dpi = 200, bbox_inches = 'tight')

plt.show()
```

```python
# 스마트폰 사용 그래프
smartphone_data = kdrama_copy2[kdrama_copy2['매체별'].str.contains('스마트폰')].set_index('매체별').T

# 70세 이상 데이터 제외
smartphone_data = smartphone_data.drop(columns='70세이상_스마트폰')

colors = ['#c2e8ff', '#38b3ff', '#89cffa', '#009eff', '#cf3232', '#ff0000']

plt.figure(figsize=(10, 5))

# 그래프 그리기
for i, col in enumerate(smartphone_data.columns):
    if i < len(colors):  # colors 리스트의 범위를 벗어나지 않도록 조건 추가
        plt.plot(smartphone_data.index[::-1], smartphone_data[col][::-1], marker='o', label=col, color=colors[i])

plt.title('연령별 스마트폰 사용')
plt.xlabel('연도')
plt.ylabel('값')
plt.legend(smartphone_data.columns, fontsize='small', loc='upper left')

plt.grid()

plt.savefig('smartphone_usage_graph.png', dpi = 200, bbox_inches = 'tight')

plt.show()
```
