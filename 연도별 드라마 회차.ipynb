```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

kdrama = pd.read_csv("kdrama.csv")
pd.set_option('display.max_rows', None)
kdrama
```

```python
# 'Year of release'와 'Number of Episodes' 컬럼만 추출
episodes_by_year = kdrama[['Year of release', 'Number of Episodes']].copy()

# 'Number of Episodes' 컬럼이 숫자형이 아니라면 숫자로 변환
episodes_by_year['Number of Episodes'] = pd.to_numeric(episodes_by_year['Number of Episodes'], errors='coerce')

# 연도별 드라마 회차 합계 계산
total_episodes_by_year = episodes_by_year.groupby('Year of release')['Number of Episodes'].sum()

# 그래프로 표시
plt.figure(figsize=(15, 8))
total_episodes_by_year.plot(kind='bar', colormap='viridis', width=0.8)
plt.title('Total Number of Episodes by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Total Number of Episodes')
plt.show()
```

```python
# 'Year of release'와 'Number of Episodes' 컬럼만 추출
episodes_by_year = kdrama[['Year of release', 'Number of Episodes']].copy()

# 'Number of Episodes' 컬럼이 숫자형이 아니라면 숫자로 변환
episodes_by_year['Number of Episodes'] = pd.to_numeric(episodes_by_year['Number of Episodes'], errors='coerce')

# 연도별 드라마 회차 평균 계산
average_episodes_by_year = episodes_by_year.groupby('Year of release')['Number of Episodes'].mean()

# 그래프로 표시
plt.figure(figsize=(15, 8))
average_episodes_by_year.plot(kind='bar', colormap='viridis', width=0.8)
plt.title('Average Number of Episodes by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Average Number of Episodes')
plt.show()
```
