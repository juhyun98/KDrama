```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
kdrama = pd.read_csv("한글제목 넣은거.csv")
kdrama
```

```python
kdrama_copy = kdrama.copy()
kdrama_copy
```

```python
kdrama_copy["Month"] = kdrama_copy["Aired Date"].str[0:3]
kdrama_copy
```

```python
kdrama_copy_mon = kdrama_copy.groupby("Month", as_index = False).agg(cnt = ("Month", "count"))

kdrama_copy_mon
```

```python
kdrama_copy_mon.plot(kind='bar', x='Month', y='cnt')
plt.xlabel('월별')
plt.ylabel('제작수')
plt.title('월별 드라마 제작수')
plt.show()
```

# ===========================================================

```python
kdrama_copy2 = kdrama.copy()
kdrama_copy2
```

```python
time_list = []

for i in range(len(kdrama_copy)) :
    time = kdrama_copy2["Duration"][i]
    
    if time[0:4] == "1 hr" :
        time_list.append(int(time[6:8]) + 60)
    elif len(time) == 12 :
        time_list.append(time[6:7])
    else :
        time_list.append(time[0:2])
        
        
kdrama_copy2["time"] = time_list

kdrama_copy2
```

```python
kdrama_copy2["time"]
```

```python
kdrama_copy2['time'] = kdrama_copy2['time'].astype(int)
kdrama_copy2
```

```python
kdrama_copy2_time = kdrama_copy2.groupby("Year of release", as_index = False).agg(time_mean = ("time", "mean"))
```

```python
kdrama_copy2_time
```

```python
kdrama_copy2_time.plot(kind='bar', x='Year of release', y='time_mean')
plt.xlabel('연도')
plt.ylabel('시간')
plt.title('연도별 드라마 시간')
plt.show()
```

# 최근 쇼츠, 릴스 등 짧은 영상 유행으로 인한 드라마 시간 조사
 
 1. 조사를 위해 한회차 드라마 시간 전처리 후 연도별 드라마 시간 평균 전처리
 2. 드라마 시간이 점점 줄어들거라는 추측과는 달리 큰 차이가 없다
 3. 왜 줄지 않는건지, 회차와 관련이 있는건지 조사 필요

```python
kdrama_copy3 = kdrama.copy()
kdrama_copy3
```

```python
kdrama_copy3_ep = kdrama_copy3.groupby("Year of release", as_index = False).agg(rat = ("Number of Episodes", "mean"))
kdrama_copy3_ep
```

```python
kdrama_copy3_ep.plot(kind='bar', x='Year of release', y='rat')
plt.xlabel('연도')
plt.ylabel('회차')
plt.title('연도별 드라마 회차')
plt.show()
```

```python
age = pd.read_csv("유료로_이용_중인_온라인동영상서비스_OTT___중복응답__20231229141028.csv", encoding = "euc-kr")
age
```

```python
age_2 = age[0:1]
age_2

age_3 = age[4:11]
age_3
```

```python
age_age = pd.concat([age_2, age_3], axis = 0)
```

```python
pd.set_option("display.max_columns", None)

age_age
```

```python
age_age_2 = age_age.rename(columns = age_age.iloc[0])

age_age_2
```

```python
age_age_2 = age_age_2.drop(index=[0])

age_age_2
```

```python
age_age_22 = age_age_2.drop('구분별(1)', axis=1)
age_age_22 = age_age_22.drop('구분별(2)', axis=1)
age_age_22 = age_age_22.drop('이용한적 없음 (%)', axis=1)

age_age_22
```

```python
age_age_22 = age_age_22.astype(float)

age_age_222 = age_age_22.sum(axis=1)

age_age_222
```

```python
age_last = pd.concat([age_age_2["구분별(2)"], age_age_222], axis = 1)

age_last
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.bar(age_last["구분별(2)"], age_last[0])

plt.title('연령대 별 OTT이용')
```
