```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

kdrama = pd.read_csv("kdrama.csv")
pd.set_option('display.max_rows', None)
kdrama
```

```python
kdrama['Genre'].unique()
```

```python
from itertools import chain

# 'Genre' 컬럼의 값을 풀어서 중복 제거
unique_genres = list(set(chain(*kdrama['Genre'].str.split(',').apply(lambda x: [genre.strip() for genre in x]))))

# 결과 출력
print(unique_genres)
len(unique_genres)
```

```python
# 'Year of release' 컬럼이 숫자형이 아니라면 숫자로 변환
kdrama['Year of release'] = pd.to_numeric(kdrama['Year of release'], errors='coerce')

# 중복 제거한 Genre 값들의 개수 계산
genre_counts = kdrama.groupby('Year of release')['Genre'].apply(lambda x: len(set(chain(*x.str.split(',').apply(lambda x: [genre.strip() for genre in x])))))

# 막대그래프로 표시
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Unique Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Unique Genres')
plt.show()
```

```python
# 'Year of release' 컬럼이 숫자형이 아니라면 숫자로 변환
kdrama['Year of release'] = pd.to_numeric(kdrama['Year of release'], errors='coerce')

# 중복 제거한 Genre 값들의 개수 계산
genre_counts = kdrama.groupby('Year of release')['Genre'].apply(lambda x: len(set(chain(*x.str.split(',').apply(lambda x: [genre.strip() for genre in x])))))

# 막대그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(genre_counts.index, genre_counts.values, color='skyblue')
plt.title('Number of Unique Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Unique Genres')
plt.show()
```

```python
# 'Year of release' 컬럼이 숫자형이 아니라면 숫자로 변환
kdrama['Year of release'] = pd.to_numeric(kdrama['Year of release'], errors='coerce')

# 각 행의 Genre 리스트에서 각 장르 개수를 세어 DataFrame 생성
genre_counts_df = pd.DataFrame()
for index, row in kdrama.iterrows():
    year = row['Year of release']
    genres = row['Genre']
    if pd.notna(genres):
        genre_list = [genre.strip() for genre in genres.split(',')]
        genre_counts = {genre: genre_list.count(genre) for genre in genre_list}
        genre_counts['Year of Release'] = year
        genre_counts_df = genre_counts_df.append(genre_counts, ignore_index=True)

# 연도별로 장르 개수 합계 계산
genre_counts_sum = genre_counts_df.groupby('Year of Release').sum()

# 표로 출력
print(genre_counts_sum)
```

```python
# 'Year of release' 컬럼이 숫자형이 아니라면 숫자로 변환
kdrama['Year of release'] = pd.to_numeric(kdrama['Year of release'], errors='coerce')

# 각 행의 Genre 리스트에서 각 장르 개수를 세어 DataFrame 생성
genre_counts_df = pd.DataFrame()
for index, row in kdrama.iterrows():
    year = row['Year of release']
    genres = row['Genre']
    if pd.notna(genres):
        genre_list = [genre.strip() for genre in genres.split(',')]
        genre_counts = {genre: genre_list.count(genre) for genre in genre_list}
        genre_counts['Year of Release'] = year
        genre_counts_df = pd.concat([genre_counts_df, pd.DataFrame([genre_counts])], ignore_index=True)

# 연도별로 장르 개수 합계 계산
genre_counts_sum = genre_counts_df.groupby('Year of Release').sum()

# 표로 출력
print(genre_counts_sum)
```

```python
# 연도별로 장르 개수 합계 계산
genre_counts_sum = genre_counts_df.groupby('Year of Release').sum()

# 막대그래프로 표시
plt.figure(figsize=(15, 8))
genre_counts_sum.plot(kind='barh', stacked=True, colormap='tab20', width=0.8)
plt.title('Number of Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Genres')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 연도별로 장르 개수 계산
genre_counts_by_year = genre_counts_df.groupby('Year of Release').sum()

# 막대그래프로 표시
plt.figure(figsize=(15, 8))

for genre in genre_counts_by_year.columns:
    plt.barh(genre_counts_by_year.index, genre_counts_by_year[genre], label=genre, alpha=0.8)

plt.title('Number of Genres by Year of Release')
plt.xlabel('Number of Genres')
plt.ylabel('Year of Release')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import seaborn as sns

# 장르별로 연도에 따른 개수 테이블 생성
genre_counts_by_year_genre = genre_counts_df.groupby(['Year of Release']).sum()

# 히트맵 생성
plt.figure(figsize=(15, 10))
sns.heatmap(genre_counts_by_year_genre.T, cmap='viridis', annot=True, fmt='g', linewidths=.5)
plt.title('Number of Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Genres')
plt.show()
```

```python
# 장르별로 연도에 따른 개수 테이블 생성
genre_counts_by_year_genre_cumsum = genre_counts_by_year_genre.cumsum()

# 누적 막대그래프로 표시
plt.figure(figsize=(15, 8))
genre_counts_by_year_genre_cumsum.plot(kind='bar', stacked=True, colormap='tab20', width=0.8)
plt.title('Cumulative Number of Genres Over the Years')
plt.xlabel('Year of Release')
plt.ylabel('Cumulative Number of Genres')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
plt.show()
```

```python
# 장르별로 연도에 따른 개수 테이블 생성
genre_counts_by_year_genre = genre_counts_df.groupby(['Year of Release']).sum()

# 그래프로 표시
plt.figure(figsize=(15, 8))
genre_counts_by_year_genre.plot(kind='bar', stacked=True, colormap='tab20', width=0.8)
plt.title('Number of Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Genres')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
plt.show()
```

```python
# 연도별로 장르 개수 합계 계산
genre_counts_sum = genre_counts_df.groupby('Year of Release').sum()

# 막대그래프로 표시
plt.figure(figsize=(15, 8))
genre_counts_sum.plot(kind='barh', stacked=True, colormap='tab20', width=0.8)
plt.title('Number of Genres by Year of Release')
plt.xlabel('Year of Release')
plt.ylabel('Number of Genres')
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
