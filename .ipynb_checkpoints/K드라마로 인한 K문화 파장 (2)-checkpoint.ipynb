```python
df = pd.read_csv('k드라마 누누티비.csv', skiprows=1, parse_dates=['주'], index_col=['주'], encoding='utf-8')

plt.figure(figsize=(15, 5))
plt.plot(df.index, df['k드라마: (전 세계)'], label='k드라마: (전 세계)')

# "2023년 4월 9일"에 해당하는 값을 찾아서 주석을 추가합니다.
peak_date = pd.Timestamp('2023-04-09')  # 날짜 포맷에 맞게 조정
peak_value = df.loc[peak_date, 'k드라마: (전 세계)']  # 해당 날짜의 값을 가져옴
plt.annotate(f'{peak_date.strftime("%Y년 %m월 %d일")}', 
             xy=(peak_date, peak_value), 
             xytext=(peak_date, peak_value-15),
             arrowprops=dict(arrowstyle='->', lw=3), 
             ha='right', color = "r",size = 18)

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()
plt.title('시간에 따른 "k드라마" 검색량',size=20)
plt.xlabel('날짜')
plt.ylabel('관심도')
plt.grid()
plt.legend()
plt.show()
```

### 로맨스 장르를 가장 많이 출연한 배우 = 김지원

```python
# 로맨스 장르 필터링
romance_dramas = kdrama[kdrama['Genre'].str.contains('Romance')]

# 'Cast' 열을 ','로 분리하여 모든 배우를 포함하는 리스트 생성
all_actors = romance_dramas['Cast'].str.split(', ', expand=True).stack().reset_index(drop=True)

# 각 배우의 출현 빈도를 계산
actor_count = all_actors.value_counts()

# 상위 1명의 배우 출력
top_actor = actor_count.head(1)
top_actor1 = kdrama[kdrama['Genre'].str.contains('Romance') & kdrama['Cast'].str.contains('Kim Ji Won')]
top_actor1
```

### 스릴러 작품을 가장 많이 쓴 작가 = 김은희

```python
thriller_dramas = kdrama[kdrama['Genre'].str.contains('Thriller')]

# 'Screenwriter' 열을 ','로 분리하여 모든 작가를 포함하는 리스트 생성
all_writers = thriller_dramas['Screenwriter'].str.split(', ', expand=True).stack().reset_index(drop=True)

# 각 작가의 출현 빈도를 계산
writer_count = all_writers.value_counts()

# 상위 1명의 작가 출력
top_writer = writer_count.head(1)
top_writer1 = kdrama[kdrama['Genre'].str.contains('Thriller') & kdrama['Screenwriter'].str.contains('Kim Eun Hee')]
top_writer1
```

### OTT 서비스의 장르별 드라마 수 

```python
ott_services = ['Netflix', 'Amazon', 'Hulu', 'Disney+']  # 원하는 OTT 서비스를 여기에 추가
ott_genre_distributions = {}

for service in ott_services:
    if service in kdrama['Original Network'].unique():
        # 해당 서비스의 데이터 필터링
        service_data = kdrama[kdrama['Original Network'].str.contains(service, na=False)]

        # 장르 데이터 추출 및 카운트
        genres = service_data['Genre'].str.split(', ', expand=True).stack().reset_index(drop=True)
        genre_count = genres.value_counts()

        # 장르 분포 저장
        ott_genre_distributions[service] = genre_count

# 시각화
fig, axes = plt.subplots(nrows=len(ott_genre_distributions), ncols=1, figsize=(10, 5 * len(ott_genre_distributions)))

if len(ott_genre_distributions) == 1:
    axes = [axes]

for i, service in enumerate(ott_genre_distributions):
    ott_genre_distributions[service].head(10).plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'{service} 장르 분포')
    axes[i].set_xlabel('장르')
    axes[i].set_ylabel('드라마 수')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
